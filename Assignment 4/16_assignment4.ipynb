```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from tqdm import tqdm
from pykeen.datasets import Nations, Kinships

# Set random seeds for reproducibility
torch.manual_seed(0)
np.random.seed(0)

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the Nations dataset using PyKEEN
nations_dataset = Nations()
nations_train_triples = nations_dataset.training.mapped_triples.tolist()
nations_test_triples = nations_dataset.testing.mapped_triples.tolist()
num_entities_nations = nations_dataset.num_entities
num_relations_nations = nations_dataset.num_relations

# Load the Kinships dataset using PyKEEN
kinships_dataset = Kinships()
kinships_train_triples = kinships_dataset.training.mapped_triples.tolist()
kinships_test_triples = kinships_dataset.testing.mapped_triples.tolist()
num_entities_kinships = kinships_dataset.num_entities
num_relations_kinships = kinships_dataset.num_relations

# TransE Model Definition
class TransE(nn.Module):
    def __init__(self, num_entities, num_relations, embedding_dim, margin, norm):
        super(TransE, self).__init__()
        self.embedding_dim = embedding_dim
        self.margin = margin
        self.norm = norm
        self.entity_embeddings = nn.Embedding(num_entities, embedding_dim)
        self.relation_embeddings = nn.Embedding(num_relations, embedding_dim)
        nn.init.uniform_(self.entity_embeddings.weight.data, -6 / np.sqrt(embedding_dim), 6 / np.sqrt(embedding_dim))
        nn.init.uniform_(self.relation_embeddings.weight.data, -6 / np.sqrt(embedding_dim), 6 / np.sqrt(embedding_dim))

    def forward(self, head, relation, tail):
        head_emb = self.entity_embeddings(head)
        rel_emb = self.relation_embeddings(relation)
        tail_emb = self.entity_embeddings(tail)
        return head_emb + rel_emb - tail_emb

    def get_triple_embedding(self, head, relation, tail):
        head_emb = self.entity_embeddings(head)
        rel_emb = self.relation_embeddings(relation)
        tail_emb = self.entity_embeddings(tail)
        return torch.cat((head_emb, rel_emb, tail_emb), dim=1)

# TransR Model Definition (similar structure as TransE)
class TransR(nn.Module):
    def __init__(self, num_entities, num_relations, embedding_dim, margin):
        super(TransR, self).__init__()
        self.embedding_dim = embedding_dim
        self.margin = margin
        self.entity_embeddings = nn.Embedding(num_entities, embedding_dim)
        self.relation_embeddings = nn.Embedding(num_relations, embedding_dim)
        self.relation_projection = nn.Embedding(num_relations, embedding_dim * embedding_dim)
        nn.init.uniform_(self.entity_embeddings.weight.data, -6 / np.sqrt(embedding_dim), 6 / np.sqrt(embedding_dim))
        nn.init.uniform_(self.relation_embeddings.weight.data, -6 / np.sqrt(embedding_dim), 6 / np.sqrt(embedding_dim))
        nn.init.uniform_(self.relation_projection.weight.data, -6 / np.sqrt(embedding_dim ** 2), 6 / np.sqrt(embedding_dim ** 2))

    def forward(self, head, relation, tail):
        head_emb = self.entity_embeddings(head)
        rel_emb = self.relation_embeddings(relation)
        tail_emb = self.entity_embeddings(tail)

        # Get the projection matrix for the relation and apply to head and tail embeddings
        projection = self.relation_projection(relation).view(-1, self.embedding_dim, self.embedding_dim)
        head_emb_projected = head_emb.view(-1, 1, self.embedding_dim) @ projection
        tail_emb_projected = tail_emb.view(-1, 1, self.embedding_dim) @ projection

        # Reshape to original embedding dimension after projection
        head_emb_projected = head_emb_projected.view(-1, self.embedding_dim)
        tail_emb_projected = tail_emb_projected.view(-1, self.embedding_dim)

        return head_emb_projected + rel_emb - tail_emb_projected

    def get_triple_embedding(self, head, relation, tail):
        head_emb = self.entity_embeddings(head)
        rel_emb = self.relation_embeddings(relation)
        tail_emb = self.entity_embeddings(tail)

        # Project both head and tail embeddings into the relation space
        projection = self.relation_projection(relation).view(-1, self.embedding_dim, self.embedding_dim)
        head_emb_projected = head_emb.view(-1, 1, self.embedding_dim) @ projection
        tail_emb_projected = tail_emb.view(-1, 1, self.embedding_dim) @ projection

        head_emb_projected = head_emb_projected.view(-1, self.embedding_dim)
        tail_emb_projected = tail_emb_projected.view(-1, self.embedding_dim)

        # Concatenate the projected embeddings with the relation embedding
        return torch.cat((head_emb_projected, rel_emb, tail_emb_projected), dim=1)


# Bernoulli Negative Sampling Function with Exclusion of Positive Samples
def bernoulli_negative_sample(triples, num_entities, num_neg_samples=1, positive_triples_set=None):
    neg_samples = []
    for h, r, t in triples:
        for _ in range(num_neg_samples):  # Generate multiple negative samples per positive triple
            while True:
                h_neg, t_neg = h, t
                if np.random.rand() < 0.5:
                    h_neg = np.random.randint(0, num_entities)
                else:
                    t_neg = np.random.randint(0, num_entities)

                # Ensure generated negative sample is not in positive_triples_set
                if positive_triples_set is None or (h_neg, r, t_neg) not in positive_triples_set:
                    neg_samples.append((h_neg, r, t_neg))
                    break  # Exit loop when a valid negative sample is found
    return neg_samples


# Margin Ranking Loss Class
class MarginRankingLoss(nn.Module):
    def __init__(self, margin):
        super(MarginRankingLoss, self).__init__()
        self.margin = margin

    def forward(self, pos_score, neg_score):
        return torch.sum(torch.clamp(self.margin + pos_score - neg_score, min=0))

import matplotlib.pyplot as plt

# Training Function with Loss Tracking
def train_model(model, optimizer, criterion, train_triples, num_entities, batch_size, num_epochs, device):
    model.train()
    losses = []
    train_triples_set = set(tuple(triple) for triple in train_triples)

    for epoch in range(num_epochs):
        epoch_loss = 0
        np.random.shuffle(train_triples)

        for i in tqdm(range(0, len(train_triples), batch_size), desc=f"Epoch {epoch + 1}"):
            batch = train_triples[i:i + batch_size]
            pos_heads, pos_rels, pos_tails = zip(*batch)

            # Convert to tensors
            pos_heads = torch.tensor(pos_heads, dtype=torch.long, device=device)
            pos_rels = torch.tensor(pos_rels, dtype=torch.long, device=device)
            pos_tails = torch.tensor(pos_tails, dtype=torch.long, device=device)

            # Generate negative samples
            neg_samples = bernoulli_negative_sample(batch, num_entities, positive_triples_set=train_triples_set)
            neg_heads, neg_rels, neg_tails = zip(*neg_samples)
            neg_heads = torch.tensor(neg_heads, dtype=torch.long, device=device)
            neg_rels = torch.tensor(neg_rels, dtype=torch.long, device=device)
            neg_tails = torch.tensor(neg_tails, dtype=torch.long, device=device)

            # Calculate positive and negative scores
            pos_scores = torch.norm(model(pos_heads, pos_rels, pos_tails), p=2, dim=1)
            neg_scores = torch.norm(model(neg_heads, neg_rels, neg_tails), p=2, dim=1)

            # Compute loss and update model
            optimizer.zero_grad()
            loss = criterion(pos_scores, neg_scores)
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        avg_loss = epoch_loss / len(train_triples)
        losses.append(avg_loss)  # Track average loss for this epoch
        print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {avg_loss}")

    return losses

def plot_loss(losses, model_name, dataset_name, margin):
    plt.figure()
    plt.plot(losses, label=f'Margin={margin}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title(f'{model_name} Training Loss on {dataset_name} (Margin={margin})')
    plt.legend()
    plt.show()

# Evaluation Function for 1-hop Question-Answering
def evaluate_model(model, test_triples, num_entities, device):
    model.eval()
    with torch.no_grad():
        head_correct = 0
        tail_correct = 0
        head_rank_sum = 0
        tail_rank_sum = 0
        hits_at_10_head = 0
        hits_at_10_tail = 0

        for head, relation, tail in test_triples:
            # Head Prediction
            head_scores = []
            for h in range(num_entities):
                score = model(torch.tensor([h], device=device), torch.tensor([relation], device=device), torch.tensor([tail], device=device))
                head_scores.append(score.norm().item())

            head_rank = np.argsort(head_scores)
            head_rank_position = np.where(head_rank == head)[0][0] + 1  # +1 for rank
            head_rank_sum += head_rank_position
            if head_rank_position <= 10:
                hits_at_10_head += 1
            if head == np.argmin(head_scores):
                head_correct += 1

            # Tail Prediction
            tail_scores = []
            for t in range(num_entities):
                score = model(torch.tensor([head], device=device), torch.tensor([relation], device=device), torch.tensor([t], device=device))
                tail_scores.append(score.norm().item())

            tail_rank = np.argsort(tail_scores)
            tail_rank_position = np.where(tail_rank == tail)[0][0] + 1  # +1 for rank
            tail_rank_sum += tail_rank_position
            if tail_rank_position <= 10:
                hits_at_10_tail += 1
            if tail == np.argmin(tail_scores):
                tail_correct += 1

        # Calculate final metrics
        mean_rank_head = head_rank_sum / len(test_triples)
        mean_rank_tail = tail_rank_sum / len(test_triples)
        hits_at_10_head_percentage = hits_at_10_head / len(test_triples)
        hits_at_10_tail_percentage = hits_at_10_tail / len(test_triples)

        return mean_rank_head, hits_at_10_head_percentage, mean_rank_tail, hits_at_10_tail_percentage

# Main Function to Train and Compare Both Models
def main():
    embedding_dim = 100
    batch_size = 128
    num_epochs = 50
    learning_rate = 0.001
    margins_nations = [1, 2, 3, 4, 5]  # For Nations dataset
    margin_kinships = 1  # For Kinships dataset

    results = {
        "TransE": {"Nations": [], "Kinships": []},
        "TransR": {"Nations": [], "Kinships": []},
    }

    # Train and evaluate TransE model on Nations dataset
    for margin in margins_nations:
        print(f"\nTraining TransE on Nations dataset with Margin: {margin}")
        transe_model = TransE(num_entities_nations, num_relations_nations, embedding_dim, margin, norm=2).to(device)
        optimizer = optim.Adam(transe_model.parameters(), lr=learning_rate)
        criterion = MarginRankingLoss(margin)
        transe_losses = train_model(transe_model, optimizer, criterion, nations_train_triples, num_entities_nations, batch_size, num_epochs, device)
        plot_loss(transe_losses, "TransE", "Nations", margin)

        # Evaluate TransE on Nations dataset
        print("Evaluating TransE on Nations dataset...")
        mean_rank_head, hits_at_10_head, mean_rank_tail, hits_at_10_tail = evaluate_model(transe_model, nations_test_triples, num_entities_nations, device)
        results["TransE"]["Nations"].append((margin, mean_rank_head, hits_at_10_head, mean_rank_tail, hits_at_10_tail))

    # Train and evaluate TransR model on Nations dataset
    for margin in margins_nations:
        print(f"\nTraining TransR on Nations dataset with Margin: {margin}")
        transr_model = TransR(num_entities_nations, num_relations_nations, embedding_dim, margin).to(device)
        optimizer = optim.Adam(transr_model.parameters(), lr=learning_rate)
        criterion = MarginRankingLoss(margin)
        transe_losses = train_model(transr_model, optimizer, criterion, nations_train_triples, num_entities_nations, batch_size, num_epochs, device)
        plot_loss(transe_losses, "TransR", "Nations", margin)

        # Evaluate TransR on Nations dataset
        print("Evaluating TransR on Nations dataset...")
        mean_rank_head, hits_at_10_head, mean_rank_tail, hits_at_10_tail = evaluate_model(transr_model, nations_test_triples, num_entities_nations, device)
        results["TransR"]["Nations"].append((margin, mean_rank_head, hits_at_10_head, mean_rank_tail, hits_at_10_tail))

    # Train and evaluate TransE model on Kinships dataset
    print(f"\nTraining TransE on Kinships dataset with Margin: {margin_kinships}")
    transe_model = TransE(num_entities_kinships, num_relations_kinships, embedding_dim, margin_kinships, norm=2).to(device)
    optimizer = optim.Adam(transe_model.parameters(), lr=learning_rate)
    criterion = MarginRankingLoss(margin_kinships)
    transe_losses = train_model(transe_model, optimizer, criterion, kinships_train_triples, num_entities_kinships, batch_size, num_epochs, device)
    plot_loss(transe_losses, "TransE", "Kinships", margin_kinships)

    # Evaluate TransE on Kinships dataset
    print("Evaluating TransE on Kinships dataset...")
    mean_rank_head, hits_at_10_head, mean_rank_tail, hits_at_10_tail = evaluate_model(transe_model, kinships_test_triples, num_entities_kinships, device)
    results["TransE"]["Kinships"].append((margin_kinships, mean_rank_head, hits_at_10_head, mean_rank_tail, hits_at_10_tail))

    # Train and evaluate TransR model on Kinships dataset
    print(f"\nTraining TransR on Kinships dataset with Margin: {margin_kinships}")
    transr_model = TransR(num_entities_kinships, num_relations_kinships, embedding_dim, margin_kinships).to(device)
    optimizer = optim.Adam(transr_model.parameters(), lr=learning_rate)
    criterion = MarginRankingLoss(margin_kinships)
    transe_losses = train_model(transr_model, optimizer, criterion, kinships_train_triples, num_entities_kinships, batch_size, num_epochs, device)
    plot_loss(transe_losses, "TransR", "Kinships", margin_kinships)

    # Evaluate TransR on Kinships dataset
    print("Evaluating TransR on Kinships dataset...")
    mean_rank_head, hits_at_10_head, mean_rank_tail, hits_at_10_tail = evaluate_model(transr_model, kinships_test_triples, num_entities_kinships, device)
    results["TransR"]["Kinships"].append((margin_kinships, mean_rank_head, hits_at_10_head, mean_rank_tail, hits_at_10_tail))

    # Print Comparison Results
    print("\nComparison Results:")
    for model in results:
        print(f"\n{model} Performance:")
        for dataset in results[model]:
            print(f"Dataset: {dataset}")
            for metrics in results[model][dataset]:
                margin = metrics[0]  # Extract the margin
                print(f"Margin: {margin}, Mean Rank (Head): {metrics[1]:.2f}, Hits@10 (Head): {metrics[2]:.2f}, Mean Rank (Tail): {metrics[3]:.2f}, Hits@10 (Tail): {metrics[4]:.2f}")

if __name__ == "__main__":
    main()
```

```python
# Set random seeds for reproducibility
torch.manual_seed(0)
np.random.seed(0)

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the Nations dataset using PyKEEN
nations_dataset = Nations()
nations_train_triples = nations_dataset.training.mapped_triples.tolist()
num_entities_nations = nations_dataset.num_entities
num_relations_nations = nations_dataset.num_relations


# Function to get cosine similarity scores
def get_cosine_similarity_scores(validation_embedding, dataset_embeddings):
    # Stack all dataset embeddings for efficient cosine similarity computation
    dataset_embeddings = torch.stack(dataset_embeddings)
    # Calculate cosine similarity between the validation embedding and each embedding in the dataset
    similarity_scores = nn.functional.cosine_similarity(
        validation_embedding.unsqueeze(0), dataset_embeddings, dim=1
    )
    return similarity_scores

# Function to retrieve top k similar triples based on cosine similarity scores
def retrieve_top_k_similar_triples(validation_embedding, dataset_embeddings, triples, k=5):
    scores = get_cosine_similarity_scores(validation_embedding, dataset_embeddings)
    scored_triples = sorted(zip(triples, scores), key=lambda x: x[1], reverse=True)[:k]
    return scored_triples

# Prepare validation triples
validation_triples = [
    ('brazil', 'commonbloc1', 'india'),
    ('burma', 'intergovorgs3', 'indonesia'),
    ('china', 'accusation', 'uk'),
    ('cuba', 'reldiplomacy', 'china'),
    ('egypt', 'embassy', 'uk')
]

# Load entity and relation mappings
entity_to_id = nations_dataset.entity_to_id
relation_to_id = nations_dataset.relation_to_id

# Convert validation triples to IDs
validation_triples_ids = [
    (entity_to_id[h], relation_to_id[r], entity_to_id[t]) for h, r, t in validation_triples
]

# Initialize models and set parameters
embedding_dim = 30
batch_size = 128
num_epochs = 50
learning_rate = 0.001

margin_transe = 3
margin_transr = 4

# Initialize models
transe_model = TransE(num_entities_nations, num_relations_nations, embedding_dim, margin=margin_transe, norm=2).to(device)
optimizer_transe = optim.Adam(transe_model.parameters(), lr=learning_rate)
criterion_transe = MarginRankingLoss(margin_transe)
transe_losses = train_model(transe_model, optimizer_transe, criterion_transe, nations_train_triples, num_entities_nations, batch_size, num_epochs, device)

transr_model = TransR(num_entities_nations, num_relations_nations, embedding_dim, margin=margin_transr).to(device)
optimizer_transr = optim.Adam(transr_model.parameters(), lr=learning_rate)
criterion_transr = MarginRankingLoss(margin_transr)
transr_losses = train_model(transr_model, optimizer_transr, criterion_transr, nations_train_triples, num_entities_nations, batch_size, num_epochs, device)

plt.figure()
plt.plot(transe_losses, label='TransE')
plt.plot(transr_losses, label='TransR')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss for TransE and TransR')
plt.legend()
plt.show()

# Extract embeddings of training triples
transe_embeddings = [
    transe_model.get_triple_embedding(torch.tensor([h], device=device), torch.tensor([r], device=device), torch.tensor([t], device=device)).squeeze()
    for h, r, t in nations_train_triples
]
transr_embeddings = [
    transr_model.get_triple_embedding(torch.tensor([h], device=device), torch.tensor([r], device=device), torch.tensor([t], device=device)).squeeze()
    for h, r, t in nations_train_triples
]

# Extract embeddings of validation triples
validation_embeddings_transe = [
    transe_model.get_triple_embedding(torch.tensor([h], device=device), torch.tensor([r], device=device), torch.tensor([t], device=device)).squeeze()
    for h, r, t in validation_triples_ids
]
validation_embeddings_transr = [
    transr_model.get_triple_embedding(torch.tensor([h], device=device), torch.tensor([r], device=device), torch.tensor([t], device=device)).squeeze()
    for h, r, t in validation_triples_ids
]

# Retrieve and print top 5 similar triples for each validation triple
entity_id_to_label = {v: k for k, v in nations_dataset.entity_to_id.items()}
relation_id_to_label = {v: k for k, v in nations_dataset.relation_to_id.items()}

for i, val_emb in enumerate(validation_embeddings_transe):
    print(f"\nValidation Triple {i+1}: {validation_triples[i]}")
    print("Top 5 similar triples using TransE embeddings with scores:")
    top_k_transe = retrieve_top_k_similar_triples(val_emb, transe_embeddings, nations_train_triples)
    for (similar_triple, score) in top_k_transe:
        triple_names = (
            entity_id_to_label[similar_triple[0]],
            relation_id_to_label[similar_triple[1]],
            entity_id_to_label[similar_triple[2]]
        )
        print(f"Triple: {triple_names}, Similarity Score: {score:.4f}")

for i, val_emb in enumerate(validation_embeddings_transr):
    print(f"\nValidation Triple {i+1}: {validation_triples[i]}")
    print("Top 5 similar triples using TransR embeddings with scores:")
    top_k_transr = retrieve_top_k_similar_triples(val_emb, transr_embeddings, nations_train_triples)
    for (similar_triple, score) in top_k_transr:
        triple_names = (
            entity_id_to_label[similar_triple[0]],
            relation_id_to_label[similar_triple[1]],
            entity_id_to_label[similar_triple[2]]
        )
        print(f"Triple: {triple_names}, Similarity Score: {score:.4f}")
```
